```python
import pandas as pd
```

```python
hdays = pd.read_csv("./Data/holidays_events.csv")
oil = pd.read_csv("./Data/oil.csv")
stores = pd.read_csv("./Data/stores.csv")
trans = pd.read_csv("./Data/transactions.csv")
train = pd.read_csv("./Data/train.csv")
test = pd.read_csv("./Data/test.csv")
submission = pd.read_csv("./Data/sample_submission.csv")
```

```python
train = pd.merge(train, oil, on = "date", how = 'left')
train = pd.merge(train, stores, on = "store_nbr", how = 'left')
train = pd.merge(train, trans, on = ["date", "store_nbr"], how = 'left')
train = pd.merge(train, hdays, on = "date", how = 'left')
```

```python
test = pd.merge(test, oil, on = "date", how = 'left')
test = pd.merge(test, stores, on = "store_nbr", how = 'left')
test = pd.merge(test, trans, on = ["date", "store_nbr"], how = 'left')
test = pd.merge(test, hdays, on = "date", how = 'left')
```
